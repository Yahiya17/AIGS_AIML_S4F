## Hierarchial Clustering

```python
# import necessary libraries 
import pandas as pd
import numpy as np 
import scipy.cluster.hierarchy as sch
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
```

```python
# loading the dataset
data = pd.read_csv("environmental_socioeconomic.csv")
data.head()
```

```python
# Normalise the features using StandardScaler
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)
print(pd.DataFrame(data_scaled,columns=data.columns).head())
```

```python
# Create a dendogram to visualize the hierarchial clustreing process
plt.figure(figsize=(10,6))
dendrogram = sch.dendrogram(sch.linkage(data_scaled,method="ward"))
plt.xlabel("Index")
plt.ylabel("Euclidean Distance")
plt.title("Dendrogram of Hierarchial Clustering")
plt.show()
```

```python
# Apply agglomerative clustering assuming 7 cluster!
hierarchial_clustering = AgglomerativeClustering(n_clusters=7,linkage="ward")
data["cluster"]=hierarchial_clustering.fit_predict(data_scaled)
```

```python
print(data.head())
```

```python
sil_score = silhouette_score(data_scaled,data["cluster"])
print(f"Sihouette Score:{sil_score}")
```

```python
# Visualize using co2 emissions and waste production
plt.figure(figsize=(8,6))
sns.scatterplot(x="co2_emissions",y="waste_production",hue="cluster",data=data,palette="viridis",s=100,alpha=0.7,edgecolor="k")
plt.title("Hierarchial Clustering of Environmental and Socioeconomic Factors")
plt.xlabel("CO2 Emissions")
plt.ylabel("Waste Production")
plt.legend(title="Cluster",bbox_to_anchor=(1.05,1),loc="upper left")
plt.show()
```
