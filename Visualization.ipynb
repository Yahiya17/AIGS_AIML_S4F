```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# line graph
x = np.arange(1,11)
y =x*2
plt.plot(x,y)
plt.xlabel("x - Axis")
plt.ylabel("Y - Axis")
plt.title("Line Plot")
plt.show()
```

```python
plt.savefig("path....png")
```

```python
plt.plot(x,y,color="r",linewidth=1,linestyle="--")
```

```python
# for 2 line graph
z = x*3
plt.plot(x,y,color="r")
plt.plot(x,z,color="b")
plt.show()
```

```python
# subplots
plt.subplot(1,2,1)
plt.plot(x,y)
plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

```python
# bar graph
names = ["sachin","dhoni","virat","yuvraj"]
runs=[78,67,87,90]
plt.bar(names,runs)
plt.show()
```

```python
plt.barh(names,runs)
```

```python
# scatter plot
x =[10,20,30,40,50,60,70,80]
y=[8,2,5,6,7,9,4,5]
plt.scatter(x,y)
plt.show()
```

```python
# Histogram
age =[1,2,3,4,5,7,8,3,7,8,9,12]
plt.hist(age,bins=20)
plt.show()
```

```python
#pie chart
company =["Apple","Samsung","LG"]
qty = [120,150,130]
plt.pie(qty,labels=company,autopct="%0.2f%%")
plt.show()
```

```python
# Box plot - outliers
x =[1,2,3,4,5,6,7,8,9]
y=[1,2,3,4,5,7,8,5,4,3,7]
z=[6,7,8,9,5,4,3]
data=list([x,y,z])
plt.boxplot(data)
plt.show()
```
